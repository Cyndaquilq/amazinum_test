1) Develop the algorithm that accepts the specified binary array of size
20x20 and iterates the steps where each step executes the following 
rules:  
● If a living cell has two or three living neighbors, it remains alive;  
● if a living cell has one or no living neighbors, it dies of  
"loneliness";   
● if a living cell has four or more living neighbors, it dies from   
"overpopulation";   
● if a dead cell has exactly three living neighbors, it comes to life.   


````python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.colors import ListedColormap
````

````python
matrix = np.random.randint(0,2,(20,20))
matrix
plt.figure(figsize=(10, 6))
cmap = ListedColormap(['slategray', 'green'])
plt.imshow(matrix, cmap=cmap, interpolation='none')
plt.show()
````

````python
def count_neighbors(grid, x, y):
   
    rows, columns = grid.shape
    count = 0                                        #Рахуємо всіх живих сусідів для елемента (x,y)    (i та  j  значення вокруг) 
    for i in range(-1, 2):
        for j in range(-1, 2):
            if i == 0 and j == 0:    #cell on skip
                continue
            sum_x = x + i
            sum_y = y + j   #sum
            if 0 <= sum_x < rows and 0 <= sum_y < columns:  # Перевірка меж
                count += grid[sum_x, sum_y]
    return count

def next_gen(current_grid):
    
    new_grid = current_grid.copy()
    rows, columns = current_grid.shape
    
    for i in range(rows):
        for j in range(columns):
            neighbors = count_neighbors(current_grid, i, j)
            
                                              
            if current_grid[i, j] == 1:                                   
                if neighbors < 2 or neighbors > 3:          
                    new_grid[i, j] = 0
            else:  
                if neighbors == 3 : 
                    new_grid[i, j] = 1
                
    return new_grid
````
